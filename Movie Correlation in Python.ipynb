```python
#Import Libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)       #adjusts the configuration of the plot that we create later

#read data
df = pd.read_csv(r'C:\Users\ANOOP\Downloads\movies.csv')
```

```python
#looking at the data

df.head()
```

```python
# check for the missing values

for col in df.columns:
    missing = np.mean(df[col].isnull())
    #print (missing)
    print('{} - {}%'.format(col, round(missing*100)))
    

```

```python
#add 0 in all the missing values 

df_new = df.fillna(0, inplace = True)
df_new
```

```python
#datatypes for our columns

df.dtypes
```

```python
#change data type of columns

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
df.head()
```

```python
df.head()
```

```python
#sort the data in descending order of gross

df = df.sort_values(by = ['gross'],ascending = False)
df.head()
```

```python
#display all rows

df_all = pd.set_option('display.max_rows',None)
```

```python
#drop any duplicates

df.drop_duplicates()
df.head()
```

```python
# Are there any Outliers?

df.boxplot(column=['gross'])
```

```python
#scatter plot with budget vs gross

plt.scatter(x='budget',y='gross',data=df,color='blue')

plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
#Budget vs gross plot using seaborn

sns.regplot(x='budget',y='gross',data=df,scatter_kws={'color':'red'},line_kws={'color':'blue'})
```

```python
#Look at correlation
```

```python
df.corr(method='pearson')     #pearson or kendall or spearman
```

```python
#Visualize correlation with a heatmap (Visualizes only the columns with numeric fields)
correlation_matrix = df.corr(method='pearson') 
sns.heatmap(correlation_matrix,annot= True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Assign numeric field values to non-numeric columns

df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name]).dtype == 'object':
        df_numerized[col_name] =  df_numerized[col_name].astype('category')
        df_numerized[col_name] =  df_numerized[col_name].cat.codes    # assingns random numeric values
        
df_numerized.head()
```

```python
# Visualize correlation of all the columns

correlation_matrix = df_numerized.corr(method='pearson') 
sns.heatmap(correlation_matrix,annot= True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized.corr()
```

```python
# Correlation in pair

corr_matrix = df_numerized.corr()
corr_pairs = corr_matrix.unstack()
corr_pairs
```

```python
# Correlation in sorted pairs

sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs)>0.5]
high_corr
```
